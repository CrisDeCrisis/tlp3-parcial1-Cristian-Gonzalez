## Importaciones necesarias

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3
```

### Acceso al dataset y creación del dataframe a través de Pandas

```python
df = pd.read_csv('votaciones.csv')
```

### 2.1 Visualisamos las primeras y ultimas filas del dataframe

```python
df.head(10)
```

```python
df.tail(10)
```

### 2.2 Obtenemos información del dataframe

```python
df.info()
```

### 2.3 Hacemos el conteo de valores nulos

```python
df.isnull().sum()
```

### 3.2 Realizamos el pasaje del tipo de dato a datetime, y luego remplazamos los valores nulos por la mediana para no perder datos.

```python
df['fecha_votacion'] = pd.to_datetime(df['fecha_votacion'], errors='coerce')

df['fecha_votacion'] = df['fecha_votacion'].fillna(df['fecha_votacion'].median())

df.head()
```

### 3.3 Capitalizamos los nombres y los apellidos

```python
df['nombre'] = df['nombre'].str.title()

df['apellido'] = df['apellido'].str.title()

df.head()
```

### 3.4 Hacemos la conversión de la columna voto para tener un formato consistente

```python
df['voto'] = df['voto'].str.upper()

df.head(10)
```

### 3.5 EL campo DNI debe ser del tipo INT.

```python
df['dni'] = pd.to_numeric(df['dni'], errors='coerce')

# df['dni'] = df['dni'].astype(int)

df.head(10)
```

### 3.6 Capitalizamos el nombre de las provincias

```python
df['provincia'] = df['provincia'].str.title()

df.head()
```

### 4.1 Calculamos los datos estadisticos de las columnas numericas

```python
df.describe().T
```

### 4.2 Calcular estadísticas específicas por grupo

```python
votos_por_provincia = df.groupby('provincia').size()

votos_por_provincia.describe()
```

### 5 Visualizamos la cantidad de votantes según la provincia

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df['provincia'], bins=60, color='tab:blue', kde=True)
plt.title('Cantidad de votantes por provincia')
plt.xticks(rotation=45)
plt.xlabel('Provincias')
plt.ylabel('Votos')
plt.grid(True)
plt.show()
```

### 6. Para finalizar, exportamos el dataframe a un archivo SQLite

```python
conn = sqlite3.connect('votaciones.sqlite')

df.to_sql('votaciones', conn, index=False)

db = pd.read_sql('SELECT * FROM votaciones', conn)

db.head(10)
```
